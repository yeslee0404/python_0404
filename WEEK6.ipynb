## 문제 1 

클래스 이름 : Lunch, Single, Combo

### 설명

Lunch, Single, Combo 클래스를 만드세요. <br>
__Single 클래스, Combo 클래스는 이미 완성된 상태입니다. 수정하지 마세요.__ <br>

Lunch 클래스의 \_\_eq\_\_, add_fries 메소드가 미완성입니다. 이 __두 메소드__ 를 완성하세요. <br>

클래스를 제대로 완성하면 아래와 같이 코드가 실행됩니다.

### 코드예시 (복사하여 테스트)

L2 = Single()  
L3 = Combo()  

print(L1==L2) # False

L1.add_fries()  
L1.add_fries()  
print(L1==L2) # True  

L2.add_fries()

print(L2==L3) # True  
print(L1.price, L2.price, L3.price) # 5000 7500 7500  
print(Lunch.price, Single.price, Combo.price) # 0 5000 7500  


```python
class Lunch:
	price = 0
	def __init__(self):
		pass
	
	def __eq__(self,p):
		return self.price==p.price
		"""두 인자의 가격(price)이 같으면 True, 다르면 False 반환 (bool type)"""

	def add_fries():
		price+=2500
		"""코드예시 확인"""

class Single(Lunch):
	price = 5000
	pass

class Combo(Lunch):
	price = 7500
	pass
```

```python
# =========================
# 문제 1: Lunch 테스트
# =========================
def test_Lunch():
    Lunch.price = 0
    Single.price = 5000
    Combo.price = 7500

    try:
        L1 = Lunch()  
        L2 = Single()  
        L3 = Combo()  
        
        results = [
            L1 == L2, # Test 1
            L2 == L3, # Test 2
            (L1.price, L2.price, L3.price) # Test 3
        ]
        
        L1.add_fries()  
        L1.add_fries()  
        results.append(L1 == L2) # Test 4
        
        L2.add_fries()  
        results.append(L1 == L2) # Test 5
        results.append(L2 == L3) # Test 6
        results.append((L1.price, L2.price, L3.price)) # Test 7
        results.append((Lunch.price, Single.price, Combo.price)) # Test 8
        
        expected = [
            False,
            False,
            (0, 5000, 7500),
            True,
            False,
            True,
            (5000, 7500, 7500),
            (0, 5000, 7500)
        ]
        
        for i in range(len(results)):
            if results[i] != expected[i]:
                print(f"❌ Lunch Test {i+1} 실패")
                print("결과:", results[i], "정답:", expected[i])
                return
        print("✅ Lunch 통과")
        
    except AttributeError as e:
        print(f"❌ Lunch 구현 오류 (메소드나 변수가 없음): {e}")
    except Exception as e:
        print(f"❌ Lunch 테스트 중 예상치 못한 에러 발생: {e}")

# =========================
# 문제 2 & Bonus: Account 테스트
# =========================
def test_Account():
    Account.rate = 5
    if 'rate' in HanaAccount.__dict__: del HanaAccount.rate
    if 'rate' in WooriAccount.__dict__: del WooriAccount.rate

    try:
        h1 = HanaAccount(0)  
        w1 = WooriAccount(10)
        
        results = [
            (h1.balance, w1.balance),        # Test 1
            (h1.rate, w1.rate)               # Test 2
        ]
        
        h1.set_rate(10)  
        w1.set_rate(15)  
        
        results.append((h1.rate, w1.rate))   # Test 3
        results.append((Account.rate, HanaAccount.rate, WooriAccount.rate)) # Test 4
        
        Account.rate = 7  
        h1.rate = 3  
        w1.rate = 1  
        results.append((Account.rate, HanaAccount.rate, WooriAccount.rate)) # Test 5
        
        # 보너스 문제 검증 부분
        global a, b
        results.append((a, b))               # Test 6
        
        expected = [
            (0, 10),
            (5, 5),
            (10, 15),
            (5, 10, 15),
            (7, 10, 15),
            (3, 10) 
        ]
        
        for i in range(len(results)):
            if results[i] != expected[i]:
                print(f"❌ Account Test {i+1} 실패")
                print("결과:", results[i], "정답:", expected[i])
                
                if i == 3: # (Account.rate, HanaAccount.rate, WooriAccount.rate) 검증
                    if results[i] == (5, 5, 5):
                        print("[피드백] set_rate에 @classmethod가 빠져있어 클래스 전체 이자율이 아닌 특정 통장(인스턴스)의 이자율만 바뀌었습니다.")
                    elif results[i] == (15, 15, 15):
                        print("[피드백] set_rate 안에서 cls.rate가 아닌 Account.rate를 하드코딩하여 부모 클래스의 값이 통째로 덮어씌워졌습니다.")
                
                return
        print("✅ Account 통과")
        
    except NameError:
        print("❌ Account Test 실패: Bonus 문제의 a, b 변수가 아직 정의되지 않았습니다.")
    except AttributeError as e:
        print(f"❌ Account 구현 오류 (메소드나 변수가 없음): {e}")
    except Exception as e:
        print(f"❌ Account 테스트 중 예상치 못한 에러 발생: {e}")


# =========================
# 문제 3: Shape 테스트
# =========================
def test_Shape():
    Shape.Vertices = 0
    Triangle.Vertices = 3
    Rectangle.Vertices = 4

    try:
        s1 = Triangle(10, 10)  
        s2 = Rectangle(10, 10) 
        
        results = [
            s1 == s2,                               # Test 1
            (s1.get_vertices(), s2.get_vertices()), # Test 2
            s1.get_area(),                          # Test 3
            s2.get_area()                           # Test 4
        ]
        
        S = Shape(10, 10)  
        S.Vertices = 4  
        results.append(S.Vertices)                  # Test 5
        results.append(S.get_vertices())            # Test 6
        
        results.append(S == s2)                     # Test 7
        
        expected = [
            False,
            (3, 4),
            50.0,
            100,
            4,
            0,
            False
        ]
        
        for i in range(len(results)):
            if results[i] != expected[i]:
                print(f"❌ Shape Test {i+1} 실패")
                print("결과:", results[i], "정답:", expected[i])
                
                # 1. 클래스 메서드 내부에서 부모 이름을 하드코딩한 경우
                if i == 1 and results[i] == (0, 0):
                    print("[피드백] get_vertices() 내에서 cls.Vertices가 아닌 Shape.Vertices를 직접 참조하여, 자식 클래스(도형)들의 꼭짓점 수가 부모의 값(0)으로 고정되었습니다.")
                # 2. 면적 값을 return하지 않은 경우
                elif i in [2, 3] and results[i] is None:
                    print("[피드백] get_area()에서 면적을 계산 후 반환(return)하지 않고 단순히 화면에 출력(print)만 한 것 같습니다.")
                # 3. @classmethod 데코레이터를 빼먹은 경우
                elif i == 5 and results[i] == 4:
                    print("[피드백] get_vertices()에 @classmethod가 없거나 내부에서 self.Vertices를 반환하여, 클래스의 원본 값이 아닌 덮어씌워진 인스턴스의 값을 가져오고 있습니다.")
                # 4. __eq__에서 type 비교를 안 한 경우
                elif i == 6 and results[i] is True:
                    print("[피드백] __eq__에서 객체의 타입(type) 자체를 비교하지 않고, 꼭짓점 수 같은 특정 속성만 비교하도록 잘못 구현되었습니다.")
                
                return
        print("✅ Shape 통과")
        
    except AttributeError as e:
        print(f"❌ Shape 구현 오류 (메소드나 변수가 없음): {e}")
    except Exception as e:
        print(f"❌ Shape 테스트 중 예상치 못한 에러 발생: {e}")
        
# 전체 테스트 실행
test_Lunch()
test_Account()
test_Shape()
```
